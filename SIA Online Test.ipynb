```python
import sys
import math
from contextlib import redirect_stdout

def compute_closest_to_zero(ts):
    if len(ts)==0 : return 0   
    
    j=0
    new = []
    while j<len(ts):
        new.append(abs(ts[j]))
        j=j+1
    print(new)
    return min(new)
    
    return -1

def main():
    n = int(input())
    ts=[int(i) for i in input().split()]
    with redirect_stdout(sys.stderr):
        solution = compute_to_zero(ts)
    print(solution)
    
if __name__=="__main__":
    main()
```
